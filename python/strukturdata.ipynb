<h2 style="background: #fff; text-align:center; padding: 12px; color:black;"><b>Struktur Data Case Hanoi</b></h2>

```python
class DiskToBig(Exception):
    """"Custom exception to handle Disc To Big then Previous disc"""
    pass
class StackUnderflowError(Exception):
    """Custom exception to handle stack underflow"""
    pass

class Stack:
    def __init__(self, name):
    
        self.stack = []
        self.name = name


    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def push(self, item):
        self.stack.append(item)
        print(f"'{item}' pushed to {self.name} stack.")


    def pop(self):
        if self.is_empty():
            raise StackUnderflowError(f"tidak bisa menghapus elemen dari stack kosong {self.name}")
        item = self.stack.pop()
        print(f"{item} popped from {self.name} stack.")
        return item
  
    def peek(self):
        return self.stack[-1]

    def display(self):
        print(f"Stack: {self.name}: {self.stack}")

def move_disk(from_stack, to_stack):
    if not from_stack.is_empty() and (to_stack.is_empty() or from_stack.peek() < to_stack.peek()):
        to_stack.push(from_stack.pop()) 
    else:
        raise DiskToBig(f"Tidak dapat memasukkan disc karena disk yang ditempatkan di atas disk yang lebih kecil.")
    

def hanoi(n,a,b,c):
    try:
        if n == 1:
            move_disk(a,b)
        else:
            hanoi(n-1, a,c,B)
            move_disk(a,b)
            hanoi(n-1, c,b,a)
    except DiskToBig as e:
        print(e)

A = Stack("A")
B = Stack("B")
C = Stack("C")



A.push(3)
A.push(2)
A.push(1)



A.display()
B.display()
C.display()

hanoi(A.size(), A, B, C)

A.display()
B.display()
C.display()
```

<h2 style="background: #fff; text-align:center; padding: 12px; color:black;"><b>Struktur Data Case Jual Beli Saham dengan LinkedList</b></h2>

```python
class Node:
    def __init__(self, qty, price):
        self.qty = qty
        self.price = price
        self.next = None

class LinkedListQueue:
    def __init__(self):
        self.next = None
        self.prev = None
        self.size = 0 

    def is_empty(self):
        return self.next is None

    def buy(self, qty, harga_beli):
        new_node = Node(qty, harga_beli)
        if self.prev is None:
            self.next = self.prev = new_node
            return
        self.prev.next = new_node
        self.prev = new_node
        self.size += 1 

    def sell(self, qty, harga_jual):
        if self.is_empty():
            return None
        
        total_gain_loss = 0
        while qty > 0 and not self.is_empty():
            temp = self.next
            if temp.qty <= qty:
                gain_loss = (harga_jual - temp.price) * temp.qty
                total_gain_loss += gain_loss
                qty -= temp.qty
                self.next = temp.next
                if self.next is None:
                    self.prev = None
                self.size -= 1 
            else:
                gain_loss = (harga_jual - temp.price) * qty
                total_gain_loss += gain_loss
                temp.qty -= qty
                qty = 0
        
        return total_gain_loss


    def display(self):
        if self.is_empty():
            print("Tidak ada antrian")
            return
        current = self.next
        while current:
            print(f"{current.qty} saham, harga {current.price}", end=" -> ")
            current = current.next
        print("None")

# Contoh penggunaan
queue = LinkedListQueue()
queue.buy(100, 20)
queue.buy(20, 24)
queue.buy(200, 36)
queue.display()
final = queue.sell(150, 30)
print(f'total keuntungan/kerugian ${final}')

```
